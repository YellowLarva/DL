```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("winequality-red.csv")
```

```python
data
```

```python
data.isna().sum()
```

```python
y=data["quality"]
X=data.drop(columns="quality")
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X= sc.fit_transform(X)
#X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state =4)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
```

```python
ann = Sequential()

# Adding the input layer and the first hidden layer
ann.add(Dense(units=11, activation='relu'))

# Adding the second hidden layer
ann.add(Dense(units=8, activation='relu'))
ann.add(Dense(units=6, activation='relu'))
ann.add(Dense(units=4, activation='relu'))
# Adding the output layer
ann.add(Dense(units=9, activation='softmax'))
# Part 3 - Training the ANN

# Compiling the ANN
import tensorflow as tf
opt = tf.keras.optimizers.Adam()
#opt = tf.train.experimental.enable_mixed_precision_graph_rewrite(opt)

ann.compile(loss="sparse_categorical_crossentropy",
              optimizer=opt,
              metrics=["accuracy"])
```

```python
ann.fit(X_train, y_train, batch_size = 64, validation_data = (X_test,y_test), epochs = 100)
```
